#### Import

```python
import numpy as np
import cv2
```

#### Read Image

```python
img = cv2.imread("temp.jpg")
```

讀進來的照片會是 `<class 'numpy.ndarray'>`

### Image Shape

```python
img.shape
```

returns a tuple，including (height, width, channel)<br/>
channel = 1: 灰階圖<br/>
channel = 3: RGB圖

### Figure Format

```python
img_gray = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)
```

cv2.IMREAD_COLOR(default): 讀取RGB三channel的彩圖，忽略透明度<br/>
cv2.IMREAD_GRAYSCALE: 灰階格式讀取圖片<br/>
cv2.IMREAD_UNCHANGED: 讀所有chanel，包含透明度

### Write Image

```python
cv2.imwrite("image.jpg", img)
```

### Show Image

```python
cv2.imshow("image.jpg", img)

# 按下任意鍵關閉所有視窗
cv2.waitKey(0)
cv2.destroyAllWindows()
```
